```python
## *args()
def pqr(a,*b,**c):
    print(f"value of a is {a},and value of b is {b} , and c is {c}")
pqr(2,5,8,name="ankit")
```

## local scope and global scope


```python
x=5
def show():
    print(x)
print(x)
show()
```

```python
x=5 ## global
def show():
    x=9 ## local 
    print(x)
print(x)
show()
```

```python
x=5
def show():
    y=9
    print(x)
    print(y)
print(x)
print(y)
show()

## 5,5,9
## 5,9,
## error
## error with output ->5,error,5,9
```

```python
x=5
def show():
    global y
    y=9
    print(x)
    print(y)
print(x)
print(y)
show()
```
