## Problem-1

## **num=[45,32,78,21,12,33,31,89,47,90,43,68]**
## find out following
## 1. all odd numbers sum
## 2. all even number sum
## 3. all divisible by 3 number sum

```python
num=[45,32,78,21,12,33,31,89,47,90,43,68]
odd=0
even=0
three_div=0
for i in num:
    if i%2!=0:
        odd=odd+i
    if i%2==0:
        even=even+i
    if i%3==0:
        three_div=three_div+i
print(f"odd {odd},even {even} , three_div {three_div}")
```

```python
def num_sum(num):
    odd=0
    even=0
    three_div=0
    for i in num:
        if i%2!=0:
            odd=odd+i
        if i%2==0:
           even=even+i
        if i%3==0:
           three_div=three_div+i
    return (odd,even,three_div)  
```

```python
l=[6,7,9,4,8]
num_sum(l)
```

```python
num=[45,32,78,21,12,33,31,89,47,90,43,68]
def num_sum(func,num):
    result=0
    for i in num:
        if func(i):
            result=result+i
    return result
x=lambda t:t%2!=0
y=lambda t:t%2==0
z=lambda t:t%3==0
print(num_sum(x,num))
print(num_sum(y,num))
print(num_sum(z,num))
     
```
