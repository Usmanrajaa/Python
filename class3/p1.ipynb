### 5. Bitwise Operators

```python
## And operator
a=1
b=1
print(a&b)
```

```python
x=2
y=3
print(x & y)
```

```python
## OR operator
j=2 ##0010
k=4 ##0100
print(j|k) ## 0110
```

```python
## Not operator
r=3
print(~r)
```

```python
## XOR Operator
# Original message
message = 5         # (binary 0101)
security_key = 3    # (binary 0011)

# Encryption
encrypted = message ^ security_key
print("Encrypted:", encrypted)

# Decryption (apply XOR again with same key)
decrypted = encrypted ^ security_key
print("Decrypted:", decrypted)
```

```python
## right and left shift operator
m=3 ## 000_0011_0000
print(m>>1)
print(m<<1)
```

### 6. Identity Operator

```python
x=5
y=5
k=5
print(id(x))
print(id(y))
print(id(k))
print(x is y)
print(x is not y)
```

```python
a=[1,2,3]
b=[1,2,3]
k=[1,2,3]
print(id(a))
print(id(b))
print(id(k))
print(a is b)
print(a is not b)
```
