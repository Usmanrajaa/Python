### Password=123456
#### Take an user input of random email , first check condition-> there must be "@" in email , if not then it won't be any mail address 
#### if there is "@" and email  found then ask the user again for correct password "123456" . if input email and the password is matching then 
#### print "welcome, enjoy your journey" but if not matched then print "password incorrect" and again ask for another correct password input . 
#### now if matches then print "finally email and password matchedd correctly"

```python
email=input("Enter your mail address:")
if "@" in email:
    print("yes it's a mail address so next you can ask for password")
    password=int(input("upar email hi hai , apna password bhi btao"))

    if email=="abcd1234@gmail.com" and password==123456:
        print("email and password correctly match kr gaye hai, enjoy!")
    elif email=="abcd1234@gmail.com" and password!=123456:
        print("incorrect password , giving you second chance")
        password=int(input("Enter your correct password again"))
        if password==123456:
            print("this time your password is correct , go and enjoy")
        else:
            print("not another chance, go to hell, try after 24 hrs") 
    else:
        print("i think user is not valid and seems fraud")         
else:
    print("given email address is not valid")
```

```python
## write a grading system that classifies the student's performance based on not only marks but 
# based on trends as well . take user input of following
# 1. Marks(0-100)
# 2. Improvements(True/False)
# 3. Attendance(0-100)
## Rules:
# Marks <40-> "Fail",
# Marks 40-59-> "pass" but if Attendance >=75 only otherwise "repeat course"
# Marks 60-70-> "good" but if Improvement="True" then "very good" otherwise "need improvement"
# Marks 70+ -> "Excellent"

# If improvement=True and marks increased from previous 
# term (simulate a random previous mark), bonus remark: "Keep it up!"
```

```python
marks=int(input("Enter marks (0-100):"))
Improvements=input("Enter (True/False)")
Attendance=int(input("Enter attendance(0-100)"))
if marks<=40:
    print("Fail")
elif marks>=40 and marks<=59:
    if Attendance>=75:
        print("pass")
    else:
        print("repeat course")
elif marks>=60 and marks<=70:
    print("good")
    if Improvements=="True":
        print(" very good")
    else:
        print("need improvement")
else:
    print("Excellent")

```
