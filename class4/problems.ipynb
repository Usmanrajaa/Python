## Problem-3

```python
## write a grading system that classifies the student's performance based on not only marks but 
# based on trends as well . take user input of following
# 1. Marks(0-100)
# 2. Improvements(True/False)
# 3. Attendance(0-100)
## Rules:
# Marks <40-> "Fail",
# Marks 40-59-> "pass" but if Attendance >=75 only otherwise "repeat course"
# Marks 60-70-> "good" but if Improvement="True" then "very good" otherwise "need improvement"
# Marks 70+ -> "Excellent"

# If improvement=True and marks increased from previous 
# term (simulate a random previous mark), bonus remark: "Keep it up!"
```

```python
marks=int(input("Enter marks (0-100):"))
Improvements=input("Enter (True/False)")
Attendance=int(input("Enter attendance(0-100)"))
if marks<=40:
    print("Fail")
elif marks>=40 and marks<=59:
    if Attendance>=75:
        print("pass")
    else:
        print("repeat course")
elif marks>=60 and marks<=70:
    print("good")
    if Improvements=="True":
        print(" very good")
    else:
        print("need improvement")
else:
    print("Excellent")

```

```python
Cart_Value = int(input("Enter cart_value: "))
is_festival = input("Enter festival status (yes/no): ")
is_first_time_customer = input("User frequency (yes for first-time/no otherwise): ")
Type_of_product = input("Product type (Grocery/Electronics): ")
referral_code = input("Referral coupon code: ")

print("\n---------- DISCOUNT DECISION LOG ----------")

# ----------------------------
# CASE 1: Cart Value <= 10000
# ----------------------------
if Cart_Value <= 10000:
    print("Cart value <= 10000 → Basic 5% discount applicable")

    if is_festival == "yes":
        print("Festival time → Additional 5% discount")
        
        if is_first_time_customer == "yes":
            print("First-time customer → Additional 5% discount")

            if Type_of_product == "Grocery":
                print("Product is Grocery → Additional 2% discount")

                if referral_code == "Mega40%" and Cart_Value >= 55000:
                    print("Referral code valid → Flat 40% discount")
                else:
                    print("No referral or not eligible for Mega40% offer")

            elif Type_of_product == "Electronics":
                print("Electronics → No additional product discount")
            else:
                print("Other product → No product-based discount")

        else:
            print("Not first-time customer → Skipping first-time discount")

    elif is_first_time_customer == "yes":
        print("Not festival, but first-time customer → Additional 5% discount")

        if Type_of_product == "Grocery":
            print("Product is Grocery → Additional 2% discount")

            if referral_code == "Mega40%" and Cart_Value >= 55000:
                print("Referral code valid → Flat 40% discount")
            else:
                print("Referral not valid or cart below limit")
        else:
            print("Product is not Grocery → No extra product discount")

    elif Type_of_product == "Grocery":
        print("Grocery item → Additional 2% discount")

        if is_festival == "yes":
            print("Festival offer also → Additional 5% discount")
            if is_first_time_customer == "yes":
                print("First-time customer → Additional 5% discount")
        else:
            print("Not festival → No extra festival offer")

    elif referral_code == "Mega40%" and Cart_Value >= 55000:
        print("Referral code valid → Flat 40% discount")
    else:
        print("Only 5% base discount applicable")

# ----------------------------
# CASE 2: 10000 < Cart Value <= 50000
# ----------------------------
elif Cart_Value > 10000 and Cart_Value <= 50000:
    print("Cart between 10000 and 50000 → Basic 10% discount applicable")

    if is_festival == "yes":
        print("Festival time → Additional 5% discount")

        if is_first_time_customer == "yes":
            print("First-time customer → Additional 5% discount")

            if Type_of_product == "Grocery":
                print("Product is Grocery → Additional 2% discount")

                if referral_code == "Mega40%" and Cart_Value >= 55000:
                    print("Referral code valid → Flat 40% discount")
                else:
                    print("No valid referral or below limit")
            else:
                print("Product not Grocery → No extra product offer")

        else:
            print("Not first-time customer → Skipping first-time discount")

    elif is_first_time_customer == "yes":
        print("First-time customer (no festival) → Additional 5% discount")

        if Type_of_product == "Grocery":
            print("Product is Grocery → Additional 2% discount")
        else:
            print("Product not Grocery → No product-based offer")

    elif Type_of_product == "Grocery":
        print("Grocery → Additional 2% discount")
    elif referral_code == "Mega40%" and Cart_Value >= 55000:
        print("Referral code valid → Flat 40% discount")
    else:
        print("Only base 10% discount applicable")

# ----------------------------
# CASE 3: Cart Value > 50000
# ----------------------------
else:
    print("Cart > 50000 → Basic flat 15% discount applicable")

    if referral_code == "Mega40%" and Cart_Value >= 55000:
        print("Referral code valid → Flat 40% discount (overrides all others)")
    else:
        if is_festival == "yes":
            print("Festival time → Additional 5% discount")

            if is_first_time_customer == "yes":
                print("First-time customer → Additional 5% discount")

                if Type_of_product == "Grocery":
                    print("Product is Grocery → Additional 2% discount")
                else:
                    print("Electronics or other → No product-based offer")
            else:
                print("Returning customer → No first-time discount")

        elif is_first_time_customer == "yes":
            print("First-time customer (no festival) → Additional 5% discount")

            if Type_of_product == "Grocery":
                print("Grocery → Additional 2% discount")
            else:
                print("Electronics → No product discount")

        elif Type_of_product == "Grocery":
            print("Grocery → Additional 2% discount")
        else:
            print("Only 15% flat discount applicable (no special condition)")

print("------------------------------------------")
print("Final applicable discounts printed above.")
```
