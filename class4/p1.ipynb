## Decision Control
### 1. if-else

```python
a=50
b=60
c=80
## Q1. if a<b and b<c then print "a<c". otherwise print any fallback message.
if a<b and b<c:
    print("a<c")
else:
    print("not matching the criteria")

```

### 2. Multiple condition with elif

```python
a=50
b=60
c=80
## Q1. if a<b and b<c then print "a<c". otherwise print any fallback message.
if a<b and b<c:
    print("a<c")
elif a<b or b>c:
    print("only a is less than b but not sure about c")
elif a<b or b<c:
    print("only b is less than c but not sure about a)")
else:
    print("not matching the criteria")

```

### 3. Nested if

```python
## take a number (>=0) and check if it is "0" then print "zero" otherwise print "positive number". 
## if not fall under any of these then  print "negative number"
```

```python
x=int(input("enter randon number"))
if x>=0:
    if x==0:
        print("zero")
    else:
        print("positive number")

else:
    print("negative number")   

```
