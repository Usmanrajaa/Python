## OOPS(object oriented programming)

```python
class Car:
    ## attributes
    def __init__(self,color,brand,switches):
        self.color=color
        self.brand=brand
        self.switches=switches
    ## methods
    def start(self):
        print(f"car will start on clicking by clicking no. of swithces {self.switches}")
        print(f"the car of this {self.color} and  brand of {self.brand} has started")

    def show(self):
        print(f"no of switches are {self.switches}")
        
car1=Car("black","mg",5) 
car2=Car("red","tata",2)
car3=Car("green","honda",3)
car1.start()
```

## 1.Encapsulation

```python
class Bankbalance:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance ## private
bank=Bankbalance("ritik",9000)

bank.balance=bank.balance+100000
bank.balance
```

```python
class Bankbalance:
    def __init__(self,owner,__balance):
        self.owner=owner
        self.__balance=__balance ## private
bank=Bankbalance("aman",9000)
bank.__balance=bank.__balance+5000
bank.__balance
```

```python
class Bankbalance:
    def __init__(self,owner,__balance):
        self.owner=owner
        self.__balance=__balance ## private
    def get(self):
        print(f"the total balance remainint is {self.__balance}")
    def set(self,amount):
        self.amount=amount
        if self.amount>0:
            self.__balance=self.__balance+amount
        else:
            print("given amount is null")

bank=Bankbalance("aman",9000)
bank.set(8000)
bank.get()
```

## 2. INHERITANCE

```python
class Bus:
    def __init__(self,brand,color,model):
        self.brand=brand
        self.color=color
        self.model=model
    def show(self):
        print(f"i have bus of this brand {self.brand} and color {self.color} and model {self.model}")
class electric_bus(Bus):
    def __init__(self,brand,color,model,battery_capacity):
        super(). __init__(brand,color,model)
        self.battery_capacity=battery_capacity
        
    def show(self):
        print(f"i have electric bus of color {self.color} and brand{self.brand} and model{self.model} battery_capacity {self.battery_capacity}")

class cng_bus(Bus):
    def __init__(self,brand,color,model,mileage):
        super(). __init__(brand,color,model)
        self.mileage=mileage
    def show(self):
        print(f"i have cng_bus of color {self.color} and brand{self.brand} and model {self.model} mileage{self.mileage}")

b1=Bus("volvo","red",2017)
b1.show()
b2=electric_bus("tata","blue",2018,"5000mAH")
b2.show()

```
