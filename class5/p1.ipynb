## 1. While loop
### executes till condition is true


```python
i=1 
while i<10: 
    print(i) 
    i+=1 
```

```python
## multiplication table of 2
i=1 
while i<10: 
    print(i*2) 
    i+=1 
```

```python
i=1 
while i<=10: 
    print("11","*",i,"=",11*i) 
    i+=1
```

```python

i=1 
j=1
while i<=1000:
    print("11 *",j,"=",i*11)
    i+=1
    j+=1
```

```python
## while loop with else
x=1
while x<3:
    print(x)
    x+=1
else:
    print("Limit exceeds")
```
