## **Problem no. 1**

### write a python program that repeatedly ask the user to enter number one by one . the program should continue accepting numbers
### until the user enters 0, which indicates the end of input
### after user enter 0 , the program should calculate and display the average of all taken numbers(excluding 0)
### if user doesn't enter any number before typing 0 then then print "no numbers entered"

```python
num=0
total_sum=0
count=0
while num>=0:
    total_sum+=num
    count+=1
    num=int(input("enter number"))
    print(num)
    if num==0:
        print("No number entered")
        break
print("total sum",total_sum)
print("total count",count)
avg=total_sum/count
print("average is ", avg)
```

## **Problem-2: Jackpot guessing game**
### write a python program that simulates random number between 1 to 100 ,
### user has to guess the number:
### after each guess if->
### 1. if guess is less than the jackpot then display "please guess higher"
### 2. if guess is higher than the jackpot  then displahy "please guess lower"
### continue guessing then number until the guess number mathces to jackpot number
### when correctly guessed then print "correct match and you won 10 lakh rupees"
### Give counter output as well


```python
import random
jackpot=random.randint(0,100)
user_input=int(input("guess the number")) 
count=0
while user_input!=jackpot:
    if user_input<jackpot:
        print("guess higher")
    else:
        print("guess lower")
    user_input=int(input("guess the number"))
    count+=1
print("you guessed in",count,"chances")
print("you won the reward")

```
