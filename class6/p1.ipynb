```python
## reverse
## range(start_value,end_value,-1), start_value>end_value
f=list(range(10,1,-1))
f
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(10,15):
    print(i)
```

```python
for i in range(15,10,-1):
    print(i)
```
