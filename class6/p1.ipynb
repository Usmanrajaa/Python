## FOR LOOP ON Containers (list, tuple, set, dictionary)

```python
l1=[2,3,4,5,6,7,8,9]
for i in l1:
    print(i)
```

```python
for i in range(len(l1)):
    print(l1[i])

```

```python
l2=[3,4,5,6,9]
for i,v in enumerate(l2):
    print(i,"-",v)
```

```python
num=int(input("enter a number:"))
for i in range(1,num+1):
    print(" "*(num-i),end="")
    for j in range(1,i):
        print(j,end='')
    for k in range(i-1,1,-1):
        print(k-1,end='')
    print()

    
```
