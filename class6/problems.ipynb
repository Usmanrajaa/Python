## ** Problem-1**
### population growth calculation
### initial population=10000
### growth rate=10%
### duration=10 years

### use loop and print each year total population value after adding the growth rate value

```python
population=10000
growth_rate=0.10
for i in range(1,11):
    population=population+population*growth_rate
    print("population after year",i,"is","-",population)
```
