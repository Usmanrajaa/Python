## ** Problem-1**
### population growth calculation
### initial population=10000
### growth rate=10%
### duration=10 years

### use loop and print each year total population value after adding the growth rate value

```python
population=10000
for i in range(1,11):
    growth_population=population*10/100
    population=population+growth_population
    print("total population in year",i,"is",population)
print("total_population after 10 years is",population)
```

## Problem no. 2
### ** factorial**
### i.e
### 1!=1*1
### 2!=2*1
### 3!=3*2*1
### 4!=4*3*2*1
### 5!=5*4*3*2*1

### take any number n by user input.

```python
num=int(input("enter any number:")) ## num=5
fact=1 ## 1,2,6,24
for i in range(1,num+1): ## 1,2,3,4,5
    fact=fact*i  ## 1*1=1, 1*2=2, 2*3=6, 6*4=24, 24*5=120
    print(fact)
print(fact)

## 5*4*3*2*1
```
